```
import numpy as np
```

```
N=50
mean=[2,-2]
cov_matrix=np.asarray([[0.9,0.2],[0.2,0.3]])
data=np.random.multivariate_normal(mean,cov_matrix,N)
print(data)
```

#the data revolves around 2,-2 ,generating 50 points.

```
N=5000
data=np.random.multivariate_normal(mean,cov_matrix,N)
print(data)
```

#using 5000 points,the data is more sampled and diverse around 2,-2
